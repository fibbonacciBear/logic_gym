````python
# Open the 1 hop dataset file

# Read problem 1

# Print
# 1. NL Problem
# 2. NL Question
# 3. Flip formalization
# 4. Flip Proof

# Manually add Flip => NL example

# This is the one shot example for GPT
````

````python
import json

file_path = "dataprep/1hop_trueontology.json"

# Load the data from the JSON file
with open(file_path, "r") as file:
    data = json.load(file)
    
````

````python
def quotes(text):
    return "```\n" + text + "\n```\n"

for index, (key, value) in enumerate(data.items()):
    markdown_text = f"# Problem {index + 1}\n\n" 
    markdown_text += "### Natural Language Problem\n"
    question = "\n".join(value["test_example"]["question"].split(". "))
    markdown_text += quotes(question)
    query = value["test_example"]["query"]
    markdown_text += "\n\n" + quotes(query)
    if "formalized" in value["test_example"]:
        markdown_text += "### Formalized"
        markdown_text += "\n\n" + quotes(value["test_example"]["formalized"])
        if "proof" in value["test_example"]:
            markdown_text += "### Formal Proof"
            markdown_text += "\n\n" + quotes(value["test_example"]["proof"])
    break

display({"text/markdown": markdown_text}, raw=True)

````
